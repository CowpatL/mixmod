# Classification, données qualitatives

Importations :

```python
from __future__ import (absolute_import, division,
                        print_function, unicode_literals) # compatibilité Python 2.7.x / 3.4.x
import pandas as pnd

import mixmod
from mixmod import gm
```

Chargement des données dans un **DataFrame** :

```python
data = pnd.read_csv(filepath_or_buffer='data/birds_txt.dat', sep=',',
                    header=False)
```

la variable contient bien un **DataFrame**:

```python
data
```

Classification avec **Mixmod** :

```python
cluster = mixmod.cluster(data,2,gm.QUALITATIVE)
```

Le résultat est une instance **ClusterOutput** :

```python
cluster
```

Affichage **"Rmixmod-like"** :

```python
print(cluster)
```

La méthode **barplot()** :

```python
%matplotlib gtk
fig = cluster.barplot(colors=['yellow', 'green'], variables=['gender',
                                                       'eyebrow'
                                                       ])
```
