# Classification, données qualitatives (2)

Importations :

```python
from __future__ import (absolute_import, division,
                        print_function, unicode_literals) # compatibilité Python 2.7.x / 3.4.x
import pandas as pnd

import mixmod
from mixmod import gm
```

Chargement des données :

```python
data = pnd.read_csv(filepath_or_buffer='data/car.csv', sep=',', header=False)
```

Classification :

```python
cluster = mixmod.cluster(data, 7, gm.QUALITATIVE)
```

L'objet résultat :

```python
cluster
```

Affichage Rmixmod-like :

```python
print(cluster)
```

La méthode **plot()** :

```python
%matplotlib gtk
fig = cluster.plot()
```
